### Rename columns

```
df1 = df[['class', 'aa_000', 'ab_000', 'ac_000', 'ad_000', 'ae_000', 'af_000']].copy()
df1.tail()
```

```
df1.info()
```

```
# To rename some columns

df1.rename({'aa_000':'temp_sensor', 'ab_000':'moist_sensor'}, axis='columns', inplace=True)

# or
#df1.rename(columns={'aa_000':'temp_sensor', 'ab_000':'moist_sensor'}, inplace=True)

df1.head()
```

```
# To rename all columns

new_cols = ['label', 'temp_sonsor', 'moist sensor', 'sensor_1', 'sensor_2', 'sensor_3', 'sensor_4']
df1.columns = new_cols
df1.head()
```

```
# To replace 'space' with '_'

df1.columns = df1.columns.str.replace(' ', '_')
df1.head()
```

```
df1.rename(str.upper, axis='columns')
```

### Remove columns

```
df2 = df.iloc[:, :8].copy()
df2.tail()
```

```
# Remove only one column

df2.drop(columns=['ag_000'], inplace=True)

# or
#df2.drop('ag_000', axis=1, inplace=True)

# or
#df2 = df2.drop(columns=['ag_000'])

df2.head()
```

```
# Remove >1 columns

df2.drop(columns=['af_000', 'aa_000'], inplace=True)
df2.head()
```

### Filter rows

```
df4 = pd.read_csv('http://bit.ly/imdbratings')
df4.sample(5)
```

```
# filter only rows in which that rows' duration are longer than 200 

df4[df4['duration'] >= 200]
```

```
# filter only rows in which that rows' rating are 'R' 

df4[df4['content_rating'] == 'R']
```

```
df4[(df4['content_rating'] == 'R') & (df4['duration'] >= 200)]
```

```
df4[(df4['genre'] == 'Biography') | ((df4['genre'] == 'Drama') & (df4.star_rating >= 8.5))]
```

```
df4[(df4['genre'] == 'Biography') | (df4['genre'] == 'Drama') | (df4['genre'] == 'Crime')]

# or
#df4[df4.genre.isin(['Crime', 'Biography', 'Drama'])]
```

### Remove duplicate rows

```
df5 = df4.sample(100, replace=True, random_state=2, axis=0)
df5
```

```
df5.duplicated().sum()
```

```
df5.loc[df5.duplicated(keep=False), :].sort_values('title')
```

```
df5.drop_duplicates(keep='first', inplace=True)
```

```
df5.loc[df5.duplicated(keep=False), :].sort_values('title')
```

### groupby

```
df6 = pd.read_csv('http://bit.ly/drinksbycountry')
df6.head()
```

```
df6.describe().T
```

```
# for each continent, what is the mean of beer servings by continent
df6.groupby('continent').beer_servings.mean()
```

```
df6[df6.continent == 'Africa'].beer_servings.mean()
```

```
df6.groupby('continent').beer_servings.describe()
```

```
df6.groupby('continent').beer_servings.agg(['count', 'min', 'max', 'mean'])
```

```
df6.groupby('continent').mean()
```

```
df6.groupby('continent').mean().plot(kind='bar');
```
