# How to manage missing values

```
import pandas as pd
```

```
df = pd.read_csv('http://bit.ly/kaggletrain')
df.head()
```

```
df.info()
```

```
df.isna().sum()
```

## Drop all missing values

```
df.dropna() #df.dropna(inplace=True)
```

```
df.dropna(axis=1)
```

```
df.isna().sum()
```

## Replace missing values

```
# แทนที่คอลัมน์ Embarked ด้วยค่า Mode

most_freq = df['Embarked'].value_counts().index[0]
df['Embarked'].fillna(value=most_freq, inplace=True)
df.isna().sum()
```

```
# แทนที่คอลัมน์ Age ด้วยค่า Mean หรือ Median

import numpy as np
mean = np.ceil(df['Age'].mean()) #median = np.ceil(df['Age'].mean())
df['Age'].fillna(value=mean, inplace=True) #df['Age'].fillna(value=median, inplace=True)
df.isna().sum()
```

```
# แทนที่คอลัมน์ Cabin ด้วยคำว่า Missing

df['Cabin'].fillna(value='missing', inplace=True)
df.isna().sum()
```
